```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
ds = pd.read_csv('../datasets/equake_clean.csv')
ds.head()
```

```python
hue_colors = { 0: "purple", 1: "red"}

sns.scatterplot(ds['age of bridge'], ds['magnitude'], data=ds, hue= 'damaged', palette = hue_colors)
plt.show()
```

```python
sns.countplot(x='bridge type', data=ds)
plt.ylabel('Count')
```

```python
ds['name'].value_counts()
```

```python
fig, ax = plt.subplots(1, 1)
ax.get_xaxis().set_visible(False)
ds['material type'].value_counts().plot(table=True, ax=ax)
```
